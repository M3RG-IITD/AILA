```python
import pandas as pd
import json
```

```python
# Load the Excel file
file_path = '/Users/indrajeetmandal/PhD/Automation/app/AFMBench/Data/database_ind.xlsx'  # Update this path with the correct file path if necessary
df = pd.read_excel(file_path)
```

```python
# Define function to convert each row to JSON format
def convert_row_to_json(row):
    json_data = {
        "question": [
            {
                "input": row['Question']
            }
        ],
        "keywords": [
            {"Operation Type":row['Operation Type'],
             "Require Tool":row['Require Tool'],
             "Require Agent":row['Require Agent'],
             "Requires":row['Requires']
            }
        ],
        "result": [
            {
                "GPT-4": row['Result GPT-4'],
                "GPT-3.5": row['Result GPT-3.5'],
            }
        ],
        "user": [row['User Name']]
    }
    return json_data

# Convert each row and save as a JSON file
for index, row in df.iterrows():
    json_data = convert_row_to_json(row)
    output_file = f'/Users/indrajeetmandal/PhD/Automation/app/AFMBench/Data/afm_qs/question_{index + 1}.json'
    with open(output_file, 'w') as json_file:
        json.dump(json_data, json_file, indent=4)
    
    print(f"JSON file saved: {output_file}")
```
